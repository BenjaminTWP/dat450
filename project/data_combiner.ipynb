```python
import pandas as pd

# File paths
file1_path = "europarl-v7.sv-en.en"
file2_path = "europarl-v7.sv-en.sv"
output_parquet = "en-sv.parquet"

# Names to use as column names in the Parquet file
file1_col_name = "en"
file2_col_name = "sv"

# Read files line by line
with open(file1_path, "r", encoding="utf-8") as f1, open(file2_path, "r", encoding="utf-8") as f2:
    lines1 = [line.strip() for line in f1.readlines()]
    lines2 = [line.strip() for line in f2.readlines()]

# Make sure both files have the same number of lines
print("len diff:", len(lines1) - len(lines2))

max_len = max(len(lines1), len(lines2))
# If one file is shorter, fill missing lines with empty strings
lines1.extend([""] * (max_len - len(lines1)))
lines2.extend([""] * (max_len - len(lines2)))

# Create a DataFrame
df = pd.DataFrame({
    file1_col_name: lines1,
    file2_col_name: lines2
})

# Save to Parquet
df.to_parquet(output_parquet, engine="pyarrow", index=False)

print(f"Parquet file saved as {output_parquet}")
```
